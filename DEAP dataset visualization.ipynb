```python
import scipy
import os
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.max_open_warning': 0})
import seaborn as sns
from scipy import io
ROOT_DIR = './'
RAW_DIR = 'data/matlabPREPROCESSED'
PROCESSED_DIR = 'data/graphProcessedData'
```

```python
colors=['#2ecc71','#3498db','#9b59b6','#c0392b']
targets = ['valence','arousal','dominance','liking']
raw_file_names = os.listdir(RAW_DIR)
raw_file_names.sort()
for name in raw_file_names:
    data = scipy.io.loadmat(f'{RAW_DIR}/{name}')
    fig, ax =plt.subplots(ncols=4, sharey=True, figsize=(20, 6))
    for i,a in enumerate(data['labels'].T):
        sns.histplot(a, kde=False, element="step", binwidth=0.5,color=colors[i],ax=ax[i]).set(xlabel=f'{targets[i]}', ylabel=f'label count({name})')
```
